  <script>
    // --- 1. Define lists ---
    let students = ["Vivian", "Sarah", "Paul"];
    let scores = [89, 20000, 78];

    let out = "";

    // --- 2. Print roster ---
    out += "<p><strong>Students:</strong> " + students.join(", ") + "</p>";

    // --- 3. Print first student and score ---
    out += "<p><strong>First student:</strong> " + students[0] + " - " + scores[0] + "</p>";

    // --- 4. Add a new student and score ---
    students.push("Dana");
    scores.push(95);
    out += "<p><strong>After adding Dana:</strong> " + students.join(", ") + "</p>";

    // --- 5. Remove a student who transfers out ---
    let index = students.indexOf("Paul"); // assume Paul leaves
    if (index !== -1) {
      students.splice(index, 1);
      scores.splice(index, 1);
    }
    out += "<p><strong>After removing Paul:</strong> " + students.join(", ") + "</p>";

    // --- 6. Curve all scores by +5 ---
    for (let i = 0; i < scores.length; i++) {
      scores[i] += 5;
    }

    // --- 7. Print each student with score ---
    out += "<h3>Final Scores</h3>";
    for (let i = 0; i < students.length; i++) {
      out += "<p>" + students[i] + " - " + scores[i] + "</p>";
    }

    // --- 8. Hack Challenge: create new list ---
    let favoriteFoods = ["Sushi", "Falafal", "Hotdogs"];
    out += "<h3>Favorite Foods</h3>";
    for (let i = 0; i < favoriteFoods.length; i++) {
      out += "<p>" + favoriteFoods[i] + "</p>";
    }

    // --- 9. Display output ---
    document.getElementById("output").innerHTML = out;
  </script>
